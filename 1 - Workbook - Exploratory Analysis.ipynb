## <span style="color:RoyalBlue"> 1.4 - Finding Correlations</span>

#### A.) Re-creating the <code style="color:steelblue">correlations</code> dataframe from <code style="color:steelblue">df</code> and visualizing the correlation grid with a basic heatmap.

```python
correlations = df.corr()
plt.figure(figsize=(7,6))
sns.heatmap(correlations, cmap='RdBu_r')
```

#### B.) Changing the background to white, reformatting map for readability.

```python
# Change color scheme
sns.set_style('white')

# Make the figsize 10 x 8
plt.figure(figsize=(10,8))

# Plot heatmap of correlations
sns.heatmap(correlations, cmap='RdBu_r')
plt.show()
```

#### C.) Displaying the correlation values in each cell.

```python
# Make the figsize 10 x 8
plt.figure(figsize=(10,8))

# Plot heatmap of annotated correlations
sns.heatmap(correlations * 100, 
            fmt='.0f', 
            annot = True, 
            cmap = 'RdBu_r')
plt.show()
```

#### D.) Printing the correlation between *beds* and *baths* and between *baths* and *beds*.


```python
df[['beds', 'baths']].corr()
```

#### E.) Generating a boolean mask for the top triangle.
* The mask should have the same shape (i.e. dimensions) as the correlation matrix.
* The values in the upper triangle should be True/1 while those in the lower triangle should be False/0.
* Either a mask of 0's and 1's OR a mask of False's and True's will work.


```python
# Generate a mask for the upper triangle
mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

#### F.) Visualizing the triangular mask to see what's going on under the hood.


```python
sns.heatmap(mask)
plt.show()
```

#### G.) Plotting the correlation heatmap again, this time using the triangular mask.


```python
# Make the figsize 10 x 8
plt.figure(figsize=(10,8))

sns.heatmap(correlations * 100, 
            fmt='.0f', 
            annot = True,
            mask = mask, 
            cmap = 'RdBu_r')
plt.show()
```

#### H.) Finally, removing the colorbar on the side.

```python
# Make the figsize 9 x 8
plt.figure(figsize=(9,8))

# Plot heatmap of correlations
sns.heatmap(correlations * 100,
            cmap='RdBu_r',
            annot=True,
            fmt='.0f',
            mask=mask,
            cbar=False)
plt.show()
```
